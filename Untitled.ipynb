```python
device = "cuda"

# -----------------------
# Paths (EDIT THESE)
# -----------------------
AE_CKPT = "/workspace/checkpoints/ae/T1/ae_patch_swin_step9000.pt"
DIFF_CKPT = "/workspace/checkpoints/diff/T1/diffusion_step_60000.pt"

# -----------------------
# Load AE
# -----------------------
enc = ShapedEncoder3D().to(device)
dec = Decoder3D().to(device)

ae_ckpt = torch.load(AE_CKPT, map_location=device)
enc.load_state_dict(ae_ckpt["encoder"])
dec.load_state_dict(ae_ckpt["decoder"])

enc.eval()
dec.eval()

for p in enc.parameters(): p.requires_grad = False
for p in dec.parameters(): p.requires_grad = False

print("✅ AE loaded")

# -----------------------
# Get latent dim
# -----------------------
with torch.no_grad():
    dummy = torch.randn(1, 1, 64, 64, 64).to(device)
    z_dummy, _ = enc(dummy)

latent_ch = z_dummy.shape[1]
print("Latent channels:", latent_ch)

# -----------------------
# Load diffusion model
# -----------------------
eps_model = ConditionalEpsUNet3D(
    z_ch=latent_ch,
    cond_ch=latent_ch
).to(device)

diff_ckpt = torch.load(DIFF_CKPT, map_location=device)
eps_model.load_state_dict(diff_ckpt["eps_model"])
eps_model.eval()

print("✅ Diffusion model loaded")

# -----------------------
# Diffusion schedule
# -----------------------
schedule = LinearNoiseSchedule(T=1000, device=device)
```

```python
class MRIDataset(torch.utils.data.Dataset):
    def __init__(self, root):
        self.files = sorted([
            os.path.join(root, f)
            for f in os.listdir(root)
            if f.endswith(".nii") or f.endswith(".nii.gz")
        ])

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        vol = nib.load(self.files[idx]).get_fdata()
        vol = torch.from_numpy(vol).float()
        vol = (vol - vol.mean()) / (vol.std() + 1e-6)  # normalize
        return vol.unsqueeze(0)  # [1, D, H, W]


```

```python
import nibabel as nib
dataset = MRIDataset("/workspace/datasets/IXI-T1")

# grab one volume
vol = dataset[0]   # [1, D, H, W], normalized

# take a random patch
x_hr = random_patch_3d(vol, 64).unsqueeze(0).to(device)

# simulate low-resolution input
x_lr = F.avg_pool3d(x_hr, kernel_size=2)

with torch.no_grad():
    z_hr, _ = enc(x_hr)
    z_lr, _ = enc(x_lr)

z_cond = F.interpolate(
    z_lr,
    size=z_hr.shape[2:],
    mode="trilinear",
    align_corners=False
)

print("Latent shape:", z_hr.shape)
```

```python
with torch.no_grad():
    z_sampled = ddim_sample(
        eps_model=eps_model,
        shape=z_hr.shape,
        schedule=schedule,
        const=schedule,
        cond=z_cond,
        steps=30,
        t_min=200
    )

print("Sampled latent:", z_sampled.shape)
```

```python
with torch.no_grad():
    z_hr, skip_hr = enc(x_hr)
    z_lr, skip_lr = enc(x_lr)

z_cond = F.interpolate(
    z_lr,
    size=z_hr.shape[2:],
    mode="trilinear",
    align_corners=False
)

with torch.no_grad():
    z_sampled = ddim_sample(
        eps_model,
        shape=z_hr.shape,
        schedule=schedule,
        const=schedule,
        cond=z_cond,
        steps=30,
        t_min=200
    )

with torch.no_grad():
    x_gen = dec(z_sampled, skip_lr)
    x_gen = match_shape(x_gen, x_hr)
```

```python
def get_mid_slices(x):
    """
    x: (1, 1, D, H, W)
    """
    D = x.shape[2]
    return x[0, 0, D // 2].cpu().numpy()

```

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_reconstructions(
    x_lr,
    x_hr,
    x_ae,
    x_ddim_skip,
    x_ddim_noskip
):
    lr = get_mid_slices(x_lr)
    hr = get_mid_slices(x_hr)
    ae = get_mid_slices(x_ae)
    ddim_s = get_mid_slices(x_ddim_skip)
    ddim_ns = get_mid_slices(x_ddim_noskip)

    err = np.abs(ddim_ns - hr)

    vmin = hr.min()
    vmax = hr.max()

    plt.figure(figsize=(18, 6))

    plt.subplot(1, 6, 1)
    plt.title("LR input")
    plt.imshow(lr, cmap="gray")
    plt.axis("off")

    plt.subplot(1, 6, 2)
    plt.title("GT HR")
    plt.imshow(hr, cmap="gray", vmin=vmin, vmax=vmax)
    plt.axis("off")

    plt.subplot(1, 6, 3)
    plt.title("AE recon")
    plt.imshow(ae, cmap="gray", vmin=vmin, vmax=vmax)
    plt.axis("off")

    plt.subplot(1, 6, 4)
    plt.title("DDIM + LR skips ❌")
    plt.imshow(ddim_s, cmap="gray", vmin=vmin, vmax=vmax)
    plt.axis("off")

    plt.subplot(1, 6, 5)
    plt.title("DDIM no skips ✅")
    plt.imshow(ddim_ns, cmap="gray", vmin=vmin, vmax=vmax)
    plt.axis("off")

    plt.subplot(1, 6, 6)
    plt.title("|Error| map")
    plt.imshow(err, cmap="hot")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```
